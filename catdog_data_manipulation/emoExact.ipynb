```python
import csv
import glob
import os
import os.path
from subprocess import call
from os import walk
import json
import random
```

```python
rawframe = []
emoClassIndex = {'행복/즐거움': 0, '편안/안정': 1, '화남/불쾌': 2, '불안/슬픔': 3, '공격성': 4, '공포': 5}
emoClass = {'행복/즐거움': [], '편안/안정': [], '화남/불쾌': [], '불안/슬픔': [], '공격성': [], '공포': []}

for subdir, dirs, files in os.walk('./'):    
    if len(subdir) > 2 and subdir[2] == '2':                           
        for file_name in files:
            if file_name[-1] == 'n':                            
                json_path = subdir[2:] + "/" + file_name                                
                dir = json_path[:-5]
                frame_count = len(glob.glob(dir + "/*.jpg"))
                with open(json_path) as json_file:
                    data = json.load(json_file)                    
                    emotion = data['inspect']['emotion']                   
                    emoIndex = emoClassIndex[emotion]                     
                    emoClass[emotion].append("{} {} {}".format(dir, frame_count, emoIndex))                   
emoClass
```

```python
split_rate = 0.8
trainRaw = {}
valRaw = {}          
arr = []
for clss in emoClass:           
    cls_len = len(emoClass[clss])
    split_point = int(split_rate * cls_len)  

    list = emoClass[clss]
    # list = [len(glob.glob(item[2:-4] + "/*.jpg")) for item in emoClass[clss]]
    # print(list)
    
    random.shuffle(list)
    trainRaw[clss] = list[:split_point]
    valRaw[clss] = list[split_point:]
    # couting each emoClass
    # arr.append([clss, len(emoClass[clss])])    
    # print('{} : {}, train: {}, val {}'.format(clss, len(emoClass[clss]), len(trainRaw[clss]), len(valRaw[clss])))            
```

```python
with open('emoTrain01.txt', 'w') as f:
    for clss in trainRaw:
        for item in trainRaw[clss]:
            f.write("%s\n" % item)
```

```python
with open('emoVal01.txt', 'w') as f:
    for clss in valRaw:
        for item in valRaw[clss]:
            f.write("%s\n" % item)
```
