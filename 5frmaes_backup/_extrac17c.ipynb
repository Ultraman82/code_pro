```python
clssInd = {}
with open('17c.txt', 'r') as fin:    
    test_list = [row.strip() for row in fin]    
    for clss in test_list:
        # print(clss.split(' '))
        [clssName, index] = clss.split(' ')
        clssInd[clssName] = index    

classes = {}
for item in clssInd:
    classes[item] = []
classes
```

```python
root_path = './'
for subdir, dirs, files in os.walk(root_path):
    for a_dir in dirs:
        if a_dir.startswith('cat') or a_dir.startswith('dog'):
            temp_class = a_dir[:-7]            
            if temp_class in classes:
                classes[temp_class].append(subdir[2:] + '/' +a_dir)
            else:
                print(a_dir)
        
```
